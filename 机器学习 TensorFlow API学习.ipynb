## 梯度下降： 

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(-6, 6,0.1)
y = x**2 +3
plt.plot(x,y)
plt.show()
```

```python
x = tf.constant([6.,0.])
for step in range(200):
    with tf.GradientTape() as tape:
        tape.watch([x])
        y = x**2+3
    grads = tape.gradient(y,[x])[0]
    x-= 0.01* grads
    if step%20==19:
        print("step {}: x = {},f(x) = {}".format(step,x.numpy(),y.numpy()))
```

## 函数拟合 

```python
x = np.linspace(0, 2*np.pi,7).reshape((-1,1))
y = tf.Variable(np.sin(x))
x = tf.Variable(x)
w1 = tf.Variable(np.random.random((1,100))*2-1)
b1 = tf.Variable(np.random.random((100,))*2-1)
w2 = tf.Variable(np.random.random((100,1))*2-1)
for step in range(100000):
    with tf.GradientTape(persistent=True) as tape:
        tape.watch([w1,b1,w2])
        out = tf.sigmoid((x@w1+b1))@w2
        loss = tf.reduce_sum(tf.square(out-y))
    grads = tape.gradient(loss, [w1,b1,w2])
    w1 = w1 - 0.01*grads[0]
    
    b1 = b1 - 0.01*grads[1]
    w2 = w2 - 0.01*grads[2]
    if step%20==19:
        print("step {}: Loss = {}".format(step,loss.numpy()))

x_test = np.linspace(0,np.pi*2,100).reshape((-1,1))
x_test = tf.constant(x_test)
y_test = tf.constant(np.sin(x_test))
y_pred = (tf.sigmoid((x_test@w1+b1))@w2).numpy()
plt.plot(x.numpy(),y.numpy(),"*")
plt.plot(x_test, y_test, x_test, y_pred)
plt.show()
```

### 二次函数拟合

```python
import numpy as np
```

```python
x = np.array(range(10)).reshape((10,1))
y = x*x + 1
```

```python
model = Sequential()
model.add(Dense(128,activation='sigmoid',input_shape=(1,)))
model.add(Dense(64,activation='sigmoid'))
model.add(Dense(32,activation='sigmoid'))
model.add(Dense(1)) 
model.compile(loss='mse',optimizer='adam')
model.fit(x,y,epochs=10000)
yp = model.predict(x)
```
